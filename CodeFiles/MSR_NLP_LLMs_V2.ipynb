```python
import pandas as pd
import re

# Load the data
df = pd.read_csv('/content/AlltimeCombined_NLPLLM_AIfroume_DataScience_Stackoverflow.csv')

# Define the tag categories and subcategories
tags_dict = {
    'Efficiency': ['efficiency', 'performance', 'optimization', 'scalability', 'latency', 'quantization', 'pruning', 'parallel-processing', 'batch-processing', 'gpu', 'tpu', 'distributed-computing', 'cloud-computing', 'concurrency', 'low-latency'],
    'Fairness': ['bias', 'ethics'],
    'Functionality': {
        'Accuracy': ['precision', 'word-completion', 'word-sense-disambiguation', 'precision-recall', 'confusion-matrix', 'precision', 'recall', 'metrics', 'auc', 'scoring', 'tf-idf', 'bleu', 'checkpoint', 'decoding', 'sampling', 'perplexity', 'accuracy', 'test-data', 'validation-datasets', 'cross-validation', 'trainingloss'],
        'Factual Correctness': ['knowledge-graph', 'relation-extraction', 'entity-linking', 'information-retrieval', 'knowledge-representation'],
        'Hallucination Detection': [],
        'Linguistic/Syntactic Correctness': ['grammar', 'parsing', 'lexical-analysis', 'pos-tagger', 'part-of-speech', 'tokenization', 'dependency-parsing', 'morphological-analysis', 'syntax', 'lemmatization', 'stemming', 'text-chunking'],
        'Reasoning Ability': ['problem-solving', 'reasoning'],
        'Semantic Correctness': ['semantics', 'semantic-analysis', 'semantic-comparison', 'context-free-grammar', 'context-sensitive-grammar', 'formal-semantics', 'inference', 'text-summarization', 'semantic-networks', 'language-agnostic', 'language-detection', 'natural-language-understanding', 'translation', 'semantic-segmentation']
    },
    'Reliability': ['versioning', 'monitoring', 'logging', 'scaling', 'domain-adaptation', 'testing', 'unit-testing', 'integration-testing', 'regression-testing', 'continuous-integration'],
    'Robustness': ['scaling', 'error-handling', 'anomaly-detection', 'data-augmentation', 'regularization', 'dropout', 'multi-task-learning'],
    'Transparency': ['explainable-ai', 'neurosymbolic-ai', 'interpretation',  'shap', 'xai', 'lime']
}

# Function to determine Category and SubCategory
def determine_category_subcategory(tag):
    for category, subcats in tags_dict.items():
        if isinstance(subcats, list):  # If subcats is a list
            if tag in subcats:
                return category, None
        elif isinstance(subcats, dict):  # If subcats is a dict
            for subcategory, tags in subcats.items():
                if tag in tags:
                    return category, subcategory
    return None, None

# Function to preprocess the tags and categorize them
def process_tags(tags):
    # Extract tags from the string using regex
    extracted_tags = re.findall(r'<(.*?)>', tags)

    categories = []
    subcategories = []

    # Determine the category and subcategory for each extracted tag
    for tag in extracted_tags:
        category, subcategory = determine_category_subcategory(tag)
        if category:
            categories.append(category)
        if subcategory:
            subcategories.append(subcategory)

    # Use 'None' as placeholder for empty categories or subcategories
    if not categories:
        categories.append('None')
    if not subcategories:
        subcategories.append('None')

    # Return the unique categories and subcategories found
    return ', '.join(set(categories)), ', '.join(set(subcategories))

# Apply the preprocessing and categorization to the Tags column
df['Category'], df['SubCategory'] = zip(*df['Tags'].apply(process_tags))

# Display the updated dataframe
print(df.head())

# Count the occurrences of each category and subcategory
category_counts = df['Category'].value_counts()
subcategory_counts = df['SubCategory'].value_counts()

print("Category Counts:\n", category_counts)
print("SubCategory Counts:\n", subcategory_counts)
```

```python

import plotly.express as px
from collections import defaultdict
import pandas as pd

# Load the CSV file
df = pd.read_csv('/content/V4_AllRecordsWCategoryWCommentCount.csv')

# Initialize defaultdict for categories and subcategories
category_counts = defaultdict(int)
subcategory_counts = defaultdict(int)

# Process categories and exclude "Functionality" and "None"
for categories in df['Category']:
    if isinstance(categories, str):
        for category in categories.split(','):
            category = category.strip()
            if category not in ['Functionality', 'None']:  # Exclude "Functionality" and "None"
                category_counts[category] += 1

# Process subcategories and exclude "Functionality" and "None"
for subcategories in df['SubCategory']:
    if isinstance(subcategories, str):
        for subcategory in subcategories.split(','):
            subcategory = subcategory.strip()
            if subcategory not in ['Functionality', 'None']:  # Exclude "Functionality" and "None"
                subcategory_counts[subcategory] += 1

# Combine the two dictionaries
combined_counts = defaultdict(int, {**category_counts, **subcategory_counts})

# Calculate the total number of occurrences
total_occurrences = sum(combined_counts.values())

# Filter out any counts with 2 or fewer occurrences and calculate percentages
filtered_combined_percentages = {task: (count / total_occurrences) * 100 for task, count in combined_counts.items() if count > 1}

# Sort the tasks by percentage
sorted_combined_tasks = sorted(filtered_combined_percentages.items(), key=lambda x: x[1], reverse=True)

# Convert to a format suitable for Plotly
tasks = [task for task, percentage in sorted_combined_tasks]
percentages = [percentage for task, percentage in sorted_combined_tasks]

# Create the Plotly bar plot (vertical orientation)
fig = px.bar(
    x=tasks,
    y=percentages,
    orientation='v',  # Vertical orientation
    labels={'y': 'Percentage of Occurrences', 'x': 'Category / SubCategory'},
    title='Distribution of Categories and Subcategories in Practice (Percentage)',
    text=[f'{p:.2f}%' for p in percentages],  # Display percentages on bars
    color=percentages,  # Add color based on percentages
    color_continuous_scale='YlGnBu'  # Use a beautiful continuous color scale
)

# Update layout for better aesthetics
fig.update_layout(
    title_font=dict(size=20, family='Arial', color='black'),
    xaxis_title_font=dict(size=16, family='Arial', color='black'),
    yaxis_title_font=dict(size=16, family='Arial', color='black'),
    yaxis=dict(
        showgrid=True,
        gridcolor='lightgrey',
        title_standoff=20,
        tickvals=[1, 10, 20, 30],  # Adjust tick values for clarity
        ticktext=['1%', '10%', '20%', '30%']  # Simplify tick labels
    ),
    xaxis=dict(showgrid=False, title_standoff=20),
    template='plotly_white',
    height=700,
    width=900,
    margin=dict(l=50, r=50, t=70, b=150)  # Adjust margins for better spacing
)

# Update the color bar (optional)
fig.update_traces(marker_line_color='black', marker_line_width=1.5, opacity=0.8)

# Show the Plotly plot
fig.show()

# Print the filtered_combined_percentages dictionary
print(filtered_combined_percentages)
```
